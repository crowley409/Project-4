```python
import pandas as pd
from scipy.stats import ttest_ind
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('clean_brain_tumor.csv')
df
```

```python
corr_matrix = df.corr()
fig, ax = plt.subplots(figsize=(15, 10))
sns.heatmap(corr_matrix,
            annot=True,
            linewidths=0.5,
            fmt=".2f")
```

```python
group_0 = df[df['Class'] == 0]['Mean']
group_1 = df[df['Class'] == 1]['Mean']
average_group_0 = group_0.mean()
average_group_1 = group_1.mean()
t_statistic, p_value = ttest_ind(group_0, group_1)
print("Averages:")
print("Group 0:")
print(average_group_0)
print("\nGroup 1:")
print(average_group_1)
print("\nT-Test Results:")
print(f'T-Statistic: {t_statistic}')
print(f'P-Value: {p_value}')
```

```python
results_df = pd.DataFrame(columns=['Feature', 'T-Statistic', 'P-Value'])
```

```python
results_list = []

features = ['Mean', 'Variance', 'Standard Deviation', 'Entropy', 'Skewness', 'Kurtosis',
            'Contrast', 'Energy', 'ASM', 'Homogeneity', 'Dissimilarity', 'Correlation']

group_0 = df[df['Class'] == 0]
group_1 = df[df['Class'] == 1]

for feature in features:
    t_statistic, p_value = ttest_ind(group_0[feature], group_1[feature])
    average_group_0 = group_0[feature].mean()
    average_group_1 = group_1[feature].mean()

    results_list.append({
        'Feature': feature,
        'T-Statistic': t_statistic,
        'P-Value': p_value,
        'Average_Group_0': average_group_0,
        'Average_Group_1': average_group_1
    })

results_df = pd.DataFrame(results_list)
results_df
```

```python
results_df.to_csv('t_test_results.csv', index=False)
```
